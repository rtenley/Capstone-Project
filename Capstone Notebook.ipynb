## Problem Statement #1

#### 1. Peoria Opioid Use / Emergency Response / Outcome Mapping

Statement / Pitch

How can we more effectively map opioid related emergency calls, ER Visits, and Deaths to better evaluate current interventions, if any, as well as identify more effective future solutions through visualizing the relationships between a number of parameters including age, geolocation, socioeconomic status, distance from medical facilities, extraneous/ambient crime, etc.?

Potential tangible outcomes:
1. Needle Exchange Site recommendations
2. Safe Use/Injection Site recommendations
3. Ideal Narcan Distribution Points/Strategies
4. Need Comprehensive Data Sharing/Dashboard Scraper
5. Identifying KPI's/Predictors of Opioid Mortality - a predictive model for prioritized tx or intervention

Corollary

Answering these questions will necessitate a review of current models, studies, and meta-analyses that may exist locally. Concurrent with seeking current and historic research in this space will be identifying and collecting the most recent datasets available from several stakeholder agencies (listed below) to help us more accurately frame hypotheses, identify knowledge gaps, and identify addtional supporting datasets. Given our current lack of knowledge, I would be most comfortable in waiting to formalize any hypotheses until we've been able to review historic analyses to inform scope. Further, once we have access to datasets, running preliminary unstructured analyses to identify potential relationships and interactions may be advised to better inform our capstone hypotheses.

    1. Directly Relevant Datasets - Expected readily available
        - Peoria GIS data
        - Peoria City/County Health Department
        - Peoria Fire Department Responses
        - Peoria Police Department Responses
        - U.S. Census
        - Illinois Bureau Labor Statistics
        
    2. Other Possible Datasets - Availabilty/Timeliness uncertain
        - Advanced Medical Transport
        - UnityPoint Health - Methodist | Proctor
        - OSF Saint Francis Medical Center
        - Illinois Department of Public Health
    
    3. Stretch Datasets - Scaling Model
        - Tazewell County Health Department
        - Woodford County Health Department
        - Tri-County Regional Planning GIS
        - Mason County Health Department
        - Logan County Health Department
        - Police / Fire for aforementioned counties
        
#### Expected Challenges

#1. Uniformity of Data

As we seek to collect data from multiple sources, we anticipate datasets that diverge widely in their collection and reporting methodoligies; we anticipate a team effort in conceptualizing our tidy-data approach and the workflow to achieve a clean, actionable data set.
         
         
#2. Lack of Context

While the opioid crisis is an exigency for our State, the lack of a shared management structure may lead to a lack of a comprehensive shared knowledge base regarding current and historic interventions and there effectiveness. Our work may help highlight this, providing us the opportunity make recommendations on better coordianting efforts.
    
    
#3. Engaged Partners

Somewhat related to the sentiments expressed in Challenge 2 above, given the disjointedness of strategic efforts in our region, it may prove more challenging to identify a clear coordinating/convening body to direct recommendations toward. With that being said, of the many problem statements we conceputualized, we feel that there would be the highest likelihood of our work on this problem to impact the practices of several discrete organizations/partners given its implications on survivability, racial equity, resource allocation, and preventitive interventions.

#4. Scope & Capacity

While this is an admirable project, we are very aware of our current individual and collective capacities regarding time available to deliver on this scope and our functional knowledge of executing a project of this magnitude. Furthermore, there may already be some community-based collaboration in this space that we could add our capacity and knowledge-base too as opposed to duplicating efforts. With that being said, we feel this is an actionable project, but more diligence is needed on current efforts to address this; we anticipate this being a focus of our group even after the class concludes.

## Problem Statement #2

#### How to better predict fires and fire risk in Peoria

Statement 

How can we more effectively identify properties at risk of fire using historic City of Peoria fire data/maps and other housing-related data sources; what are the precedents of fire and can we build a model to help predict where fires may be most likely to occur in neighborhoods of high risk?

Potential tangible outcomes:
1. Examining and extrapolating data from metro firemaps to support response times/reources of neighborhood fire houses
2. Building a predictive model based on a number of parameters including home age, location, code vialtion history, etc.
3. More effective prioritizing of fire department resources/preventative measures
4. More effective prioritizing of inspections

Corollary

To our knowledge, Peoria does not currently have a mechanism or process, outside of routine inspections which have a substantial backlog. Thus, with our current connection to the City of Peoria Fire Department through this project, and tagentially the City's Code Enforcement Department, we have ready access to all the data we need. Thus, this is an actionable project that will have tangible outcomes that will help priortize/expedite inspections in backlog and more effectively allocate resources. Different from the abstract nature of our opioid problem statement, this problem seems to have more concrete hypothetical precedents that we can begin testing as soon as data is cleaned; that is, we may not need to spend as much time running exploratory analyses to hone in on precedents, interactions, and relationships.
